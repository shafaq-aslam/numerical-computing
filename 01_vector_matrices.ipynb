# 1.1 Vector Representation

```python
class Vector:
    def __init__(self, x=0.0, y=0.0):
        self.x = x
        self.y = y

    def __str__(self):
        return "[{}, {}]".format(str(self.x), str(self.y))
```

```python
v1 = Vector(2, 8)
print(v1)
```

```python
v2 = Vector(5, 3)
print(v2)
```

```python
def add(self, v2):
    v3 = Vector()
    v3.x = self.x + v2.x
    v3.y = self.y + v2.y
    return v3

Vector.add = add
```

```python
z = v1.add(v2)
print(z)
```

```python
def mul(self, s):
    return Vector(s * self.x, s * self.y)

Vector.mul = mul
```

```python
a = v1.mul(5)
print(a)
```

```python
def sub(self, v2):
    return self.add(v2.mul(-1))

Vector.sub = sub
```

```python
a_min_v2 = a.sub(v2)
print(a_min_v2)
```
